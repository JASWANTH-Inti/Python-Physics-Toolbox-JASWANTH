```python
# Task 1 – Plasma Temperature Classifier
# To check the category of Plasma based on input data

def classify_plasma_temp(temp):                              # Temp in Kelvin
    if temp < 10000:
        return "Cold plasma"
    elif 10000 <= temp <= 1000000:
        return "Warm plasma"
    else:
        return "Hot plasma"

results = []

for i in range(5):
    while True:
        try:
            t = float(input(f"Enter temperature {i+1} in Kelvin: "))
            category = classify_plasma_temp(t)
            results.append((t, category))
            break
        except ValueError:
            print("Invalid input. Please enter a number.")

print("Plasma Temperature Classification:")
print("--------------------------------")
print("Temperature (K) Category")
for temp, cat in results:
    print(temp, cat)
```

# Task 1 Summary
This code gives a simple function to classify plasma based on temperature into cold, warm, and hot plasma types. 
The program takes temperature inputs from the user, handles invalid input, and stores the results. 

```python
# Task 2 – Material Properties Lookup
# Comparing the properties of elements, and extracting useful physical parameters like density and electrical conductivity for analysis.

materials = {
    "Aluminum": {"density_g_cm3": 2.7, "conductivity": 35000000, "youngs_modulus_GPa": 70},
    "Copper": {"density_g_cm3": 8.96, "conductivity": 59000000, "youngs_modulus_GPa": 110},
    "Silicon": {"density_g_cm3": 2.33, "conductivity": 0.0001, "youngs_modulus_GPa": 130},
    "Graphene": {"density_g_cm3": 2.2, "conductivity": 100000000, "youngs_modulus_GPa": 1000}
}

def add_material(name, density, conductivity, youngs_modulus):
    materials[name] = {
        "density_g_cm3": density,                    # density in g/cm^3
        "conductivity": conductivity,                # Conductivity in S/m
        "youngs_modulus_GPa": youngs_modulus         # Youngs modulus in Giga Pa (10^9) 
    }

def highest_conductivity_material():
    max_material = ""
    max_conductivity = 0

    for name in materials:
        cond = materials[name]["conductivity"]
        if cond > max_conductivity:
            max_conductivity = cond
            max_material = name

    return max_material

def get_densities():
    densities = []
    for name in materials:
        densities.append(materials[name]["density_g_cm3"])
    return densities

# Example usage
add_material("Iron", 7.87, 10000000, 200)

print("Material with highest conductivity:", highest_conductivity_material())
print("All densities:", get_densities())
```

# Task 2 Summary
This code creates a small materials database using Python dictionaries to store physical properties like density, electrical conductivity, and Young’s modulus. 
Functions are used to add new materials, find the material with the highest conductivity, and extract densities for analysis. 

```python
# Task 3 – Particle Energy Calculator
# This program uses a Particle class to store particle properties and calculate kinetic energy, which helps model simple physical systems using OOPS
#Each particle is represented as an object with mass and velocity. 

class Particle:
    def __init__(self, name, mass_kg, velocity_m_s):
        self.name = name
        self.mass_kg = mass_kg
        self.velocity_m_s = velocity_m_s

    def kinetic_energy(self):
        energy = 0.5 * self.mass_kg * self.velocity_m_s * self.velocity_m_s
        return energy

# Create particle objects
electron = Particle("Electron", 9.11e-31, 20000000)
proton = Particle("Proton", 1.67e-27, 1000000)
alpha = Particle("Alpha particle", 6.64e-27, 500000)

particles = [electron, proton, alpha]

print("Particle Energies:")
print("------------------")

for p in particles:
    energy_joule = p.kinetic_energy()
    energy_ev = energy_joule / 1.6e-19
    print(p.name, ":", energy_ev, "eV")
```

# Task 3 Summary
This code uses a simple Particle class to demonstrate physical particles with mass and velocity. 
We define a method to compute the kinetic energy of each particle, and the energies are converted into electron-volts for recognition.
