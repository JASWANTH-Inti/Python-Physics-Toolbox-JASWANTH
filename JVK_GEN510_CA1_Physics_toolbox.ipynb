# Task 1 – Plasma Temperature Classifier (Loops + Conditionals + Functions)
Write a function classify_plasma_temp(temp) that takes temperature (float, in Kelvin) and returns:
•	"Cold plasma" (< 10^4 K)
•	"Warm plasma" (10^4 – 10^6 K)
•	"Hot plasma" (> 10^6 K)
Use it in a loop: ask user for 5 temperatures (handle invalid input with try-except), classify each, and store results in a list of tuples (temp, category). Print nicely formatted table. Add 1–2 sentences: why temperature classification matters in plasma physics experiments.


```python
# Task 1 – Plasma Temperature Classifier
# To check the category of Plasma based on input data

data = []

def classify_plasma_temp(temp):                              # Temp in Kelvin
    if temp < 10000:
        return "Cold plasma"
    elif 10000 <= temp <= 1000000:
        return "Warm plasma"
    else:
        return "Hot plasma"
    
for i in range(5):
    while True:
        try:
            t = float(input("Enter temp {i+1} in k: "))
            category = classify_plasma_temp(t)
            data.append((t, category))
            break
        except ValueError:
            print("Error. Enter a number.")

print("Plasma Temp Classification:")
print("Temperature (K) Category")
for temp, cat in data:
    print(temp, cat)
```

# Task 1 Summary
This code gives a simple function to classify plasma based on temperature into cold, warm, and hot plasma types. 
The program takes temperature inputs from the user, handles invalid input, and stores the results. 

# Task 2 – Material Properties Lookup (Dictionaries + Lists + Strings)
Create a dictionary with 4–5 materials from materials science (e.g. Aluminum, Copper, Silicon, Graphene). Each value = another dict: {'density_g_cm3': 2.7, 'conductivity': 3.5e7, 'youngs_modulus_GPa': 70}
Write functions to:
•	Add a new material
•	Find material with highest conductivity
•	Return list of densities as floats (use list comprehension + string cleaning if needed)
Show example usage + short comment on how dicts help organize experimental data.


```python
# Task 2 – Material Properties Lookup
# Comparing the properties of elements, and extracting useful physical parameters like density and electrical conductivity for analysis.

materials = {
    "Aluminum": {"density_g/cm3": 2.7, "conductivity_S/m": 35000000, "youngs_modulus_GPa": 70},
    "Copper": {"density_g/cm3": 8.96, "conductivity_S/m": 59000000, "youngs_modulus_GPa": 110},
    "Silicon": {"density_g/cm3": 2.33, "conductivity_S/m": 0.0001, "youngs_modulus_GPa": 130},
    "Graphene": {"density_g/cm3": 2.2, "conductivity_S/m": 100000000, "youngs_modulus_GPa": 1000}
}

def add_material(element, density, conductivity, youngs_modulus):
    materials[element] = {
        "density_g/cm3": density,                    # density in g/cm^3
        "conductivity_S/m": conductivity,                # Conductivity in S/m
        "youngs_modulus_GPa": youngs_modulus         # Youngs modulus in Giga Pa (10^9) 
    }

def highest_conductivity_material():
    max_material = ""
    max_conductivity = 0

    for element in materials:
        cond = materials[element]["conductivity_S/m"]
        if cond > max_conductivity:
            max_conductivity = cond
            max_material = element

    return max_material

def get_densities():
    densities = []
    for element in materials:
        densities.append(materials[element]["density_g/cm3"])
    return densities

# Ex. Usage
add_material("Iron", 7.87, 10000000, 200)

print("Material with highest conductivity:", highest_conductivity_material())
print("All densities:", get_densities())
```

# Task 2 Summary
Using python dictionaries, this code generates a small material database to save physical properties such as density, electrical conductivity, and Young’s modulus. 
Functions are used to add new materials, find materials with the highest conductivity, and extract densities for references and analysis. 

# Task 3 – Simple Particle Energy Calculator (Functions + OOP basics)
Define a class Particle with attributes: mass_kg, velocity_m_s, name (str)
Add method kinetic_energy() → return ½mv² (use float)
Create 3 particle objects (e.g. electron, proton, alpha particle with realistic masses/velocities). Loop over them → print name + energy in eV (convert using 1 eV ≈ 1.6e-19 J).
Bonus (extra 2 marks if time): add charge attribute and simple potential_energy(q, V) method.


```python
# Task 3 – Particle Energy Calculator
# This program uses a Particle class to store particle properties and calculate kinetic energy, which helps model simple physical systems using OOPS
# Each particle is represented as an object with mass and velocity. 

class Particle:
    def __init__(self, name, mass_kg, velocity_m_s):
        self.name = name
        self.mass_kg = mass_kg
        self.velocity_m_s = velocity_m_s

    def kinetic_energy(self):
        energy = 0.5 * self.mass_kg * self.velocity_m_s * self.velocity_m_s
        return energy

# Creating particle objects
electron = Particle("Electron", 9.11e-31, 20000000)
proton = Particle("Proton", 1.67e-27, 1000000)
alpha = Particle("Alpha particle", 6.64e-27, 500000)

particles = [electron, proton, alpha]

print("Particle Energies:")

for p in particles:
    energy_joule = p.kinetic_energy()
    energy_ev = energy_joule / 1.6e-19
    print(p.name, ":", energy_ev, "eV")
```

# Task 3 Summary
This code uses a simple Particle class to display physical particles with mass and velocity. 
We define a technique to compute the kinetic energy of each particle, and the energies are converted into electron-volts for recognition.
